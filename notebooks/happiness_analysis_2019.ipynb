```python
# Create the histogram
plt.figure(figsize=(10, 6))
ax = sns.histplot(df['Score'], kde=True)

# Label the axes and add a title
plt.xlabel("Score")
plt.ylabel("Number of Countries")
plt.title("Distribution of Happiness Scores (2019)")

# Add Numerical labels to the bars
for p in ax. patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')


# removes yticklabels
ax.set_yticklabels([])
ax.set_yticks([])
# removes right and top spines
plt.gca().spines[['right', 'top']].set_visible(False)
# Show the plot
plt.show()
```

```python
df_sorted = df.sort_values(by='Score', ascending=False) # Sorting Rows based on Score
top_10_countries = df_sorted.head(10) # Selecting top 10 countries

print("Top 10 Countries in 2019:")
for index, row in top_10_countries.iterrows():
    print(f"{row['Country or region']}: {row['Score']}")
```

```python
df_sorted = df.sort_values(by='Score', ascending=True)
bottom_10_countries = df_sorted.head(10)

print("Bottom 10 least happiest countries in 2019:")
for index, row in bottom_10_countries.iterrows():
  print(f"{row['Country or region']}: {row['Score']}")
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='Score', y='Country or region', data=top_10_countries, palette= "Blues")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Happiness Score")
plt.ylabel("")
plt.title("Top 10 Happiest Countries in 2019")

for p in ax.patches:
    ax.annotate(f'{p.get_width():.2f}', (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha = 'left', va = 'center', xytext = (5, 0), textcoords = 'offset points')


ax.set_xticklabels([])
ax.set_xticks([])
plt.gca().spines[['right', 'top']].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot between Score vs GDP per capita
plt.figure(figsize=(10, 6))
sns.regplot(x='GDP per capita', y='Score', data=df, ci=None)
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.title("Happiness vs GDP per Capita with Trend Line (2019)")
plt.gca().spines[['right', 'top']].set_visible(False)
plt.show()
```

```python
# Heat map to find correlation among columns
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation Matrix of Numeric Columns (2019)")
plt.show()
```
